```
r, cnt = [[0]*6, [0]*6], 0
n, k = map(int, input().split())
for i in range(n):
    s, y = map(int, input().split())
    r[s][y-1] += 1
for i in r:
    for j in i:
        if j != 0:
            cnt += (j-1)//k+1
print(cnt)
```

## Silver III
---
문제 번호 : 18429

제목 : [근손실](https://www.acmicpc.net/problem/18429)

배점 : 8점

```
import itertools as it
n, k = map(int, input().split())
a = input().split()
kit = {str(i+1): int(a[i]) for i in range(n)}
p = list(it.permutations(kit))
cnt = 0
for i in p:
    w = 0
    for j in i:
        w += kit[j] - k
        if w < 0:
            cnt += 1
            break
print(len(p)-cnt)
```

## Silver II
---
문제 번호 : 15888

제목 : [정답은 이수근이야!](https://www.acmicpc.net/problem/15888)

배점 : 9점

```
a, b, c = map(int, input().split())
d = b**2-4*a*c 
if d <= 0:
    print("둘다틀렸근")
else:
    root1 = (-b+d**.5)/(2*a)
    root2 = (-b-d**.5)/(2*a)
    if root1-int(root1) == 0 and root2-int(root2) == 0:
        root1, root2 = int(root1), int(root2)
        if (not root1 & (root1-1)) and (not root2 & (root2-1)) and root1 > 1 and root2 > 1:
            print("이수근")
        else:
            print("정수근")
    else:
        print("둘다틀렸근")
```

## Gold IV
---
문제 번호 : 1339

제목 : [단어 수학](https://www.acmicpc.net/problem/1339)

배점 : 12점

```
alpha = [0 for i in range(26)]
for i in range(int(input())):
    str1 = input()
    for i in range(len(str1)):
        alpha[ord(str1[i])-65] += 10 ** (len(str1)-i-1)
alpha.sort(reverse=True)
for i in range(26):
    alpha[i] *= (9-i)
    if alpha[i] == 0:
        break
print(sum(alpha))
```

## Gold III
---
문제 번호 : 11054

제목 : [가장 긴 바이토닉 부분 수열](https://www.acmicpc.net/problem/11054)

배점 : 13점

```
n = int(input())
arr = list(map(int, input().split()))
dp_ = [1 for i in range(n)]
dp_r = [1 for i in range(n)]
for i in range(n):
    for j in range(i):
        if arr[i] > arr[j]:
            dp_[i] = max(dp_[i], dp_[j] + 1)
arr.reverse()
for i in range(n):
    for j in range(i):
        if arr[i] > arr[j]:
            dp_r[i] = max(dp_r[i], dp_r[j] + 1)
dp_r.reverse()
a = [0 for i in range(n)]
for i in range(n):
    a[i] = dp_[i] + dp_r[i]
print(max(a) - 1)
```

## Gold I
---
문제 번호 : 1016

제목 : [제곱 ㄴㄴ 수](https://www.acmicpc.net/problem/1016)

배점 : 15점

```
from math import ceil

a, b = map(int, input().split())
r = [0 for i in range(a, b + 1)]
s = [i ** 2 for i in range(2, 1000001)]
n = 2
for i in s:
    if a // i * i < a: num = (a // i + 1) * i - a
    else: num = ceil(a / i) * i - a
    while num <= b - a:
        r[num] = 1
        num += i
result = 0
for i in r:
    if i == 0: result += 1
print(result)
```
